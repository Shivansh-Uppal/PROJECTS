```python
ap=pd.get_dummies(data['sex'])
data=pd.concat([data,ap],axis=1)
```

```python
rm=pd.get_dummies(data['romantic'])
data=pd.concat([data,rm],axis=1)
```

```python
ny=pd.get_dummies(data['nursery'])
data=pd.concat([data,ny],axis=1)
```

```python
hi=pd.get_dummies(data['higher'])
data=pd.concat([data,hi],axis=1)
```

```python
fa=pd.get_dummies(data['famsup'])
data=pd.concat([data,fa],axis=1)
```

```python
f=pd.get_dummies(data['famsize'])
data=pd.concat([data,f],axis=1)
```

```python
ps=pd.get_dummies(data['Pstatus'])
data=pd.concat([data,ps],axis=1)
```

```python
data.drop(['paid','romantic','activities','nursery','higher','internet','schoolsup','famsup','sex','famsize','Pstatus','guardian'],
          inplace=True,axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['G1','G2','G3'],axis=1)
y=data[['G1','G2','G3']]
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
from sklearn import tree
```

```python
clf=tree.DecisionTreeRegressor()
```

```python
clf=clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput import MultiOutputRegressor
```

```python
clf2=make_pipeline(StandardScaler(),SVR(C=1,epsilon=0.2))
clf2=MultiOutputRegressor(clf2)
```

```python
clf2.fit(X_train,y_train)
```

```python
clf2.predict(X_test)
```
